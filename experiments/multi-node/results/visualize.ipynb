```python
import sys
import glob
import os
import re
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import datetime
%matplotlib inline
```

```python
noseqr = pd.read_table("bench-ng.csv", sep=(","))
seqr = pd.read_table("bench-ng-local.csv", sep=(","))
append_noseqr = pd.read_table("bench-append-ng.csv", sep=(","))
append_seqr = pd.read_table("bench-ng-append-local.csv", sep=(","))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(seqr['avg_iops'], 'rx-', markersize=5, linewidth=1.0, label="co-located client/sequencer")
ax.plot(noseqr['avg_iops'], 'b.-', markersize=10, linewidth=1.0, label="remote sequencer")
ax.plot(append_seqr['avg_iops'], 'g+-', markersize=10, linewidth=1.0, label="appends co-located")
ax.plot(append_noseqr['avg_iops'], 'y.-', markersize=10, linewidth=1.0, label="appends remote sequencer")

ax.set_ylim(0, 400)
ax.set_xlim(0, 20)
ax.set_ylabel("Average IOPs")
ax.set_xlabel("Tiempo (secs)")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(seqr['avg_lat'], 'rx-', linewidth=1.0, markersize=5, label="co-located client/sequencer")
ax.plot(noseqr['avg_lat'], 'b.-', linewidth=1.0, markersize=10, label="remote sequencer")
ax.plot(append_seqr['avg_lat'], 'g+-', markersize=10, linewidth=1.0, label="appends co-located")
ax.plot(append_noseqr['avg_lat'], 'y.-', markersize=10, linewidth=1.0, label="appends remote sequencer")

ax.set_ylim(bottom=0)
ax.set_xlim(0, 20)
ax.set_ylabel("Latency (ms)")
ax.set_xlabel("Tiempo (secs)")
lgd = ax.legend(loc='center', bbox_to_anchor=(1.5, 0.8))
```
